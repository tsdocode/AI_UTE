$$2x^8 + 5x^7 -3x^5 + 4x^5 - 2.5x^4 + 11x^3 -2x +5 = 0$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Khảo sát hàm số

```python
def func(x):
    return 2*x**8+5*x**7-3*x**6+4*x**5-2.5*x**4+11*x**3-2*x+5
```

```python
x = np.linspace(-30, 30, 100)
```

```python
plt.plot(x, func(x))
```

# Tìm nghiệm

```python
def solve(alpha, beta, i , f):
    xrange = np.arange(i,f,beta)
    y = func(xrange)
    index = np.argmin(np.abs(y))
    rs = [xrange[index], y[index]]
    if (rs[1] < alpha):
        return rs
    else:
        return "Chua tim thay nghiem yeu cau"
```

```python
rs = solve(0.01, 0.0001, -1 , 1)
```

```python
plt.plot(x, func(x))
plt.scatter(rs[0], rs[1], color = 'red')
```
